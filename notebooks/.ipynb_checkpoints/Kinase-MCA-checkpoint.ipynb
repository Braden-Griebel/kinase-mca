# Multiple Correspondance Analysis (MCA) of Kinase Targets in *Mycobacterium tuberculosis*

This analysis with Multiple Correspondence Analysis (MCA) to evaluate if there are groups of genes targeted by a simmilar group of kinases, as well as determining simmilarities between the kinases based on which genes they target. 

## 1. Import Needed Libraries

```python
# Standard Library Imports
from itertools import combinations
import os
from typing import Callable
import warnings

# External Imports
import matplotlib.pyplot as plt
import numpy as np
import optuna
import pandas as pd
import prince
from scipy.cluster.hierarchy import dendrogram
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN, OPTICS
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.exceptions import ConvergenceWarning

# Local Imports
```

## 2. Load and Preprocess Data

Data for the phosphorylation sites from:  
Frando, A., Boradia, V., Gritsenko, M. et al. The Mycobacterium tuberculosis protein O-phosphorylation landscape. Nat Microbiol 8, 548â€“561 (2023). [https://doi.org/10.1038/s41564-022-01313-7](https://www.nature.com/articles/s41564-022-01313-7)

Data has sheets for each of the kinases studied, which includes which columns for:  
 - Mutant: Whether the phosphorylation was observed based on an OE or LOF mutant
 - Peptide: Peptide tag from the Mass Spectrometry Experiments
 - Rv Number: The locus tag for a gene
 - Gene Name: The name for the gene (Blank if no name assigned)
 - Phosphosite: Where on the gene the kinase phosphorylated, in the form of \<Rv Number\>_\<AA Number\>_\<AA Residue\>
 - Quantification of the amount of phosphorylation and significance level
 - Indirect?: Whether the phosphorylation effect is considered to be indirect

```python
# Read the Mtb genes data to get a list of all genes
gene_info = pd.read_csv(os.path.join("..", "data", "Mtb_genes.csv"))
gene_list = pd.Index(gene_info["Locus"].values)
kinase_list = ["PknB","PknD","PknE","PknF","PknG","PknH","PknI","PknJ","PknK","PknL"]

direct_phos_df = pd.DataFrame(0, index=gene_list, columns = kinase_list)
total_phos_df = pd.DataFrame(0, index=gene_list, columns = kinase_list)
```

```python
# Create a dictionary of name conversions, to convert the Frando genes to Mycobrowser genes
name_conv_dict = {"Rv1155a": "Rv1155", "Rv2975a":"Rv2975c"}

# For each kinase, read the correct sheet and get a list of the phosphorylated genes
for kinase in kinase_list:
    df = pd.read_excel(os.path.join("..","data", "frando_phosphosite_data.xlsx"), 
                       sheet_name=kinase).replace(name_conv_dict)
    direct_df = df[~(df["Indirect?"]=="Indirect")]

    total_phos_df.loc[df["Rv Number"], kinase] = 1
    direct_phos_df.loc[direct_df["Rv Number"], kinase] = 1

# Now only keep the rows with at least 1 phosphorylation (TODO: VALID?)
direct_phos_df = direct_phos_df[direct_phos_df.sum(axis=1)>0]
total_phos_df = total_phos_df[total_phos_df.sum(axis=1)>0]
```

## 3. Perform MCA Decomposition
Use the Prince package to perform MCA on the kinase targeting data

### 3a. Fit initial MCA

```python
# MCA
mca_res = prince.MCA(n_components = 10, one_hot=False, check_input=True)
mca_res = mca_res.fit(direct_phos_df)
```

### 3b. Plot Cumulative % of Variance explained

```python
fig, ax = plt.subplots()

# Convert cumulative % of Variance explained to float
eig_sum = mca_res.eigenvalues_summary
eig_sum["cumulative explained"] = eig_sum["% of variance (cumulative)"].str.rstrip("%").astype(float)

# Plot the results
sns.lineplot(eig_sum, x=mca_res.eigenvalues_summary.index, y="cumulative explained")
ax.set_title("Cumulative % Variance Explained")
ax.set_ylabel("Cumulative Variance Explained")
ax.set_xlabel("Number of Components")

ax.hlines(y=70, xmin=0, xmax=9, colors="r")
```

As can be seen in the above graph, with 6 components, >70% of the variance can be explained, so 6 components will be used for the remainder of the analysis. 

```python
# Set N_COMPONENTS based on above graph
N_COMPONENTS=6

# Compute the final MCA with this number of components
mca_final = prince.MCA(n_components = N_COMPONENTS, one_hot=False, check_input=True)
mca_final = mca_final.fit(direct_phos_df)
```

### 3c. Plot the MCA results

```python
mca_final.plot(direct_phos_df, x_component=0, y_component=1,
            show_column_markers=True, 
            show_column_labels=True, 
            show_row_labels=False)
```

### 4. Dendrogram Plotting Function

```python
# Function from https://scikit-learn.org/stable/auto_examples/cluster/plot_agglomerative_dendrogram.html
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

### 5. Weighted Euclidean Distance Agglomerative Clustering

```python
# Functions for finding the weighted euclidean distance 
def weighted_eucl_dist(x:np.array,y:np.array,weights:np.array)->float:
    """
    Function to determine the weighted Euclidean Distance between two points
    Params
    ------
    x : np.array
        First Point
    y : np.array
        Second Point
    weights : np.array
        Weights
    Returns
    -------
    float
        The weighted distance
    """
    return np.sqrt(np.multiply(np.power(x-y,2), weights).sum())

def weighted_eucl_dist_mat(X:np.array, weights:np.array)->np.array:
    """
    Function to find the Weighted Euclidean distance matrix
    Params
    ------
    X : np.array
        Array of points for which to compute the distance matrix, columns representing dimensions, 
        rows representing points
    weights : np.array
        Weights for each dimension, should have len(weight) == X.shape[1]
    Returns
    -------
    np.array
        The weighted distance matrix, dimensions are (nxn) where n is the number of rows in X
    """
    if len(weights)!=X.shape[1]:
        raise ValueError("Weights must be the same length as the number of columns in X")
    distance_matrix = np.zeros((X.shape[0], X.shape[0]), float)
    for i,j in combinations(range(0,X.shape[0]),2):
        distance_matrix[i,j] = weighted_eucl_dist(X[i], X[j], weights)
    return distance_matrix+distance_matrix.T
```

```python
# Choose Number of components
N_WEIGHTED_COMPONENTS = 9

# Compute the MCA for the weighted clustering (the MCA isn't weighted)
mca_weighted = prince.MCA(n_components = N_WEIGHTED_COMPONENTS, one_hot=False, check_input=True)
mca_weighted = mca_weighted.fit(total_phos_df)#direct_phos_df)



# Get the columns coordinates, and select number of columns
w_col_coords = mca_weighted.column_coordinates(total_phos_df).values
# Select the number of eigenvalues
eigenvalue_weights = mca_weighted.eigenvalues_
# Calculate the weighted distance matrix
w_col_distance = weighted_eucl_dist_mat(w_col_coords, eigenvalue_weights)
```

```python
# Cluster using the weighted distance matrix
weighted_agg_clustering = AgglomerativeClustering(metric="precomputed", linkage="average", n_clusters=None, distance_threshold=0)
weighted_agg_clustering = weighted_agg_clustering.fit(w_col_distance)
plt.title("Weighted Hierarchical Clustering of Kinases")
plot_dendrogram(weighted_agg_clustering, labels = kinase_list)
```

## 5. Gene (row) Clustering

Now the genes will be clustered to determine which genes are phosphorylated by simmilar groups of kinases. To find the best clustering, KMeans, DBSCAN, and OPTICS will be evaluated (with their respective hyperparameters) using the silhouette score. To improve the search speed the optuna library implementing a Parzen Tree structured estimator will be used to find the optimal model, and hyperparameters. 

```python
# Get the row coordinates
row_coords = mca_final.row_coordinates(direct_phos_df)
# Since the row coordinates of some of the genes are identical (i.e. those that are targeted by exactly the same kinases)
# This will fi
row_coords_duplicates = row_coords.groupby(row_coords.columns.tolist(), as_index=False).size()
unique_points = len(row_coords_duplicates)
```

```python
# Plot the gene dendrogram 
agg_cluster = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

agg_cluster = agg_cluster.fit(row_coords)

plt.figure(figsize=(100,50))
plt.title("Hierarchical Clustering of Genes")
plot_dendrogram(agg_cluster, labels = row_coords.index)
plt.savefig("gene_clustering.png")
```

### 5.1
Evaluating hyperparameters and clustering methods

### 5.1a KMeans

```python
# For each possible cluster size 
n_clusters = np.arange(2, unique_points)
silhouette = np.zeros(len(n_clusters)) # to store silhouette scores
inertia = np.zeros(len(n_clusters)) # to store inertia
kmeans = KMeans() # Initialize KMeans object
for index, nc in enumerate(n_clusters):
    kmeans.set_params(n_clusters = nc)
    clusters = kmeans.fit_predict(row_coords)
    inertia[index] = kmeans.inertia_
    silhouette[index] = silhouette_score(row_coords, clusters)
```

```python
fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(n_clusters, inertia)
ax[0].set_title("Inertia Plot")
ax[0].set_ylabel("Inertia")


ax[1].plot(n_clusters, silhouette)
ax[1].set_title("Silhouette Coefficient")
ax[1].set_xlabel("Number of Clusters")
ax[1].set_ylabel("Silhouette Coefficient")
```

### 5.1b Agglomerative Clustering

```python
agglomerative_clustering = AgglomerativeClustering()
def agg_objective(trial):
    # Suggest number of cluseters to use
    linkage = trial.suggest_categorical("linkage", ["ward", "complete", "average", "single"])
    distance_or_n_clusters = trial.suggest_categorical("distance_or_n_clusters", ["distance", "n_clusters"])
    if distance_or_n_clusters == "distance":
        distance_threshold = trial.suggest_float("distance_threshold", low=0.001, high=10)
        n_clusters=None
    else:
        n_clusters = trial.suggest_int("n_clusters", low=2, high=unique_points)
        distance_threshold=None
    agglomerative_clustering.set_params(n_clusters = n_clusters, linkage=linkage, distance_threshold=distance_threshold)
    clusters = agglomerative_clustering.fit_predict(row_coords)
    if len(np.unique(clusters)) < 2:
        return -1
    return silhouette_score(row_coords, clusters)
```

```python
agg_sampler = optuna.samplers.TPESampler(n_startup_trials=100)
agg_study = optuna.create_study(study_name = "Agglomerative Clustering Optimization", direction="maximize", sampler=agg_sampler)
optuna.logging.set_verbosity(optuna.logging.ERROR)
agg_study.optimize(agg_objective, n_trials=1000, show_progress_bar=True)
```

```python
print(f"The best parameters were found in trial {agg_study.best_trial.number} and were {agg_study.best_params}")
```
